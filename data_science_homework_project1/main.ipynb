```python
import os
import csv
import json
from collections import namedtuple
```

```python
#Q1: What are the names of the files present in the data directory?
def list_files_in(pathname):
    files_in_path = sorted(os.listdir(pathname))
    name_no_dot = list()
    for name in files_in_path:
        if name[0] != ".":
            name_no_dot.append(name)
    
    
    return name_no_dot
files_name_in_data = list_files_in("data")
files_name_in_data
```

```python
#Q2: What are the paths of all the files in the data directory?
def list_paths_in(pathname):
    files_join_list = list()
    for file in list_files_in(pathname):
        file_join = os.path.join(pathname, file)
        files_join_list.append(file_join)
    return files_join_list
data_paths = list_paths_in("data")
data_paths
```

```python
#Q3: What are the paths of all the files whose name contains comment_data in our data directory?
comment_paths_temp = list()
comment_paths = list()
for paths in data_paths:
    if "comment_data" in paths:
        comment_paths_temp.append(paths)
    comment_paths = sorted(comment_paths_temp)
comment_paths        
```

```python
#Q5: What is the name of the channel with the ID UCNye-wNBqNL5ZzHSJj3l8Bg?
def read_json(path):
    with open(path, encoding="utf-8") as f:
        return json.load(f)
    
def get_mapping(pathname):
    try:
        json_data = read_json(pathname)
        return json_data  
    except (json.JSONDecodeError) as e:
        empty_dict = {}
        return empty_dict

get_mapping(os.path.join("data", "channel_ids1.json"))["UCNye-wNBqNL5ZzHSJj3l8Bg"]

        
```

```python
#Q7: What is the Comment object with comment ID UgygOezB4Mvd5o6FgAt4AaABAg?

Comment = namedtuple("Comment", ["video_id", "comment_length", "author_id", "likes", "published_at"])

def process_csv(filename):
    exampleFile = open(filename, encoding="utf-8")  
    exampleReader = csv.reader(exampleFile)
    exampleData = list(exampleReader)        
    exampleFile.close()  
    return exampleData

comment_data_dict = {}
def get_comment_data(comment_file):
    csv_data = process_csv(comment_file)
    header = csv_data[0]
    comment_rows = csv_data[1:]
    comment_id_idx = header.index("comment_id")
    video_id_idx = header.index("video_id")
    comment_length_idx = header.index("comment_length")
    author_id_idx = header.index("author_id")
    likes_idx = header.index("likes")
    published_at_idx = header.index("published_at")
    for rows in comment_rows:
        if "" in rows:
            continue
        else:
            try:
                c = Comment(rows[video_id_idx], int(rows[comment_length_idx]), rows[author_id_idx], int(rows[likes_idx]), rows[published_at_idx])
                comment_data_dict[rows[comment_id_idx]] = c
            except:
                continue
    return comment_data_dict       


get_comment_data(os.path.join("data", "comment_data1.csv"))["UgygOezB4Mvd5o6FgAt4AaABAg"]
```

```python
#Q13: What is the video with ID fkMW60W180E?
def get_videos(data_file, video_mapping_file):
    data = process_csv(data_file)
    header = data[0]
    all_videos = data[1:]
    video_mapping = get_mapping(video_mapping_file)
    videos_dict = dict()
    for video in all_videos: # You may find it helpful to do all_videos[:5] to only look at the first 5 videos.
        video_ID = video[header.index("video_id")]
        for key in video_mapping:
            if video_ID == key:
                if video_ID not in videos_dict:
                    videos_dict[video_ID]={}
                    videos_dict[video_ID]["title"]=video_mapping[video_ID]
                    videos_dict[video_ID]["channel_name"] = channel_dict[video[header.index("channel_id")]]
                    videos_dict[video_ID]["comments"] = comment_buckets[video_ID]
                for headers in header:
                    if headers == "channel_id" or headers == "video_id":
                        continue
                    elif headers not in videos_dict[video_ID]:
                        videos_dict[video_ID][headers]= None
                    if headers == "views":
                        videos_dict[video_ID][headers] = int(video[header.index(headers)])
                    elif headers == "likes":
                        if video[header.index(headers)] == "":
                            videos_dict[video_ID]["ratings_enabled"] = False
                            videos_dict[video_ID][headers] = None
                        else:
                            videos_dict[video_ID]["ratings_enabled"] = True
                            videos_dict[video_ID][headers] = int(video[header.index(headers)]) 
                    elif headers == "dislikes":
                        if video[header.index(headers)] == "":
                            videos_dict[video_ID]["ratings_enabled"] = False
                            videos_dict[video_ID][headers] = None
                        else:
                            videos_dict[video_ID]["ratings_enabled"] = True 
                            videos_dict[video_ID][headers] = int(video[header.index(headers)])
                    elif headers == "tags":
                        videos_dict[video_ID][headers] = (video[header.index(headers)].split("|"))
                    else:
                        videos_dict[video_ID][headers]=video[header.index(headers)]
    return videos_dict

videos = get_videos(os.path.join('data','video_data.csv'), os.path.join('data','video_ids.json'))

sample = ['UgzgwN2JXxjTN4mR5954AaABAg.9TPxukUd20g9TQLnJi3RFU', 'UgzvogxMg82Kj0aW84x4AaABAg']
for s in sample:
    assert s in videos['fkMW60W180E']['comments']
videos["fkMW60W180E"]
```

```python
#Q17: Which video titles were produced by the "Corridor Crew" channel?
def bucketize(attribute, videos=videos):
    attribute_dict={}
    for video in videos:
        for key in videos[video]:
            if key == attribute:
                if type(videos[video][key]) == list:
                    for item in videos[video][key]:
                        if item not in attribute_dict:
                            attribute_dict[item]=[]
                        attribute_dict[item].append(video)
                else:
                    if videos[video][key] not in attribute_dict:
                        attribute_dict[videos[video][key]] = []
                    attribute_dict[videos[video][key]].append(video)
                    
    return attribute_dict

category_dict = bucketize('category')
assert category_dict['Pets & Animals'] == ['Hz_DslzN2IA', 'AwvyrO_yM4c']

channel_buckets = bucketize("channel_name")
video_id_list_cc = channel_buckets["Corridor Crew"]
[videos[video_id]["title"] for video_id in video_id_list_cc]
```

```python
#Q18: What are the top 5 channels that have the most total comments on their videos?
channel_comments_dict = {}

for channel in channel_buckets:
    if channel not in channel_comments_dict:
        channel_comments_dict[channel]= None 
    curr_comments = 0
    for video in channel_buckets[channel]:
        curr_comments += len(videos[video]["comments"])
        channel_comments_dict[channel]=curr_comments

channel_comments_rank = dict(sorted(channel_comments_dict.items(), key = lambda t:t[1], reverse = True))

channel_rank_list = list()
for channel_name in channel_comments_rank:
    channel_rank_list.append(channel_name)
channel_rank_list[0:5]
```

```python
#Q19: List all the unique video titles which have Minecraft tags.
tag_buckets = bucketize("tags")
video_title_list = list()
video_title_list_temp = list()
minecraft = "Minecraft"
for tag in tag_buckets:
    if minecraft.lower() in tag.lower():
        for video_id in tag_buckets[tag]:
            video_title = videos[video_id]["title"]
            video_title_list_temp.append(video_title)
video_title_list = list(set(video_title_list_temp))
video_title_list
```
