```python
import numpy as np
```

```python
# Build a single neuron (two-inputs)
# define activation function
def sigmoid(x):
    return 1/(1+np.exp(-x))

# create class Neuron
class Neuron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias
    def feedforward(self, inputs):
        # put weighted inputs & bias into activation function
        total = np.dot(self.weights, inputs) + self.bias
        return sigmoid(total)

weights = np.array([1, 1]) # w1 = 1, w2 = 1
bias = 1     # b = 1
n = Neuron(weights, bias)

x = np.array([1, 2]) # x1 = 1, x2 = 2
print(n.feedforward(x))
        
```

```python
# Neural network with 2 hidden neurons:
class NeuralNet:
    def __init__(self):
        weights = np.array([0,1]) 
        bias = 0
        self.h1 = Neuron(weights, bias)
        self.h2 = Neuron(weights, bias)
        self.o1 = Neuron(weights, bias)
    def feedforward(self,x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        out_o1 = self.o1.feedforward(np.array([out_h1,out_h2]))
        return out_o1

network1 = NeuralNet()
x = np.array([2,3])
print(network1.feedforward(x)) 
```

```python
# Build a neural net
# A neural net with 2 inputs and 3 neurons in 1 hidden layer,
# and an output layer with 1 neuron. 
class NeuralNet1:
    def __init__(self):
        weights = np.array([0,1]) # weights for 2 inputs
        weights2 = np.array([0,1,1]) # weights for outputs of three hidden neurons. Required three elements for three neurons
        bias = 0
        self.h1 = Neuron(weights, bias)
        self.h2 = Neuron(weights, bias)
        self.h3 = Neuron(weights, bias)
        self.o1 = Neuron(weights2, bias)
    def feedforward(self,x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        out_h3 = self.h3.feedforward(x)
        out_o1 = self.o1.feedforward(np.array([out_h1,out_h2, out_h3]))
        return out_o1

network1 = NeuralNet1()
x = np.array([2,3])
print(network1.feedforward(x)) 
```

```python
# Build a neural net
# A neural net with 2 inputs and 4 neurons in 1 hidden layer,
# and an output layer with 1 neuron. 
class NeuralNet2:
    def __init__(self):
        weights = np.array([0,1]) # weights for 2 inputs
        weights3 = np.array([0,1,1,1]) # weights for outputs of 4 hidden neurons. Required 4 elements for 4 neurons
        bias = 0
        self.h1 = Neuron(weights, bias)
        self.h2 = Neuron(weights, bias)
        self.h3 = Neuron(weights, bias)
        self.h4 = Neuron(weights,bias)
        self.o1 = Neuron(weights3, bias)
    def feedforward(self,x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        out_h3 = self.h3.feedforward(x)
        out_h4 = self.h4.feedforward(x)
        out_o1 = self.o1.feedforward(np.array([out_h1,out_h2, out_h3, out_h4]))
        return out_o1

network1 = NeuralNet2()
x = np.array([2,3])
print(network1.feedforward(x)) 
```

```python
# Build a neural net
# A neural net with 2 inputs and 4 neurons in 1 hidden layer,
# and an output layer with 1 neuron. 
# different weights and biases between neurons.
class NeuralNet3:
    def __init__(self):
        weights = np.array([0,1]) 
        weights3 = np.array([0,1,1,1]) # weights for outputs of 4 hidden neurons. Required 4 elements for 4 neurons
        weights4 = np.array([2,5])
        weights5 = np.array([17,18])
        weights6 = np.array([0.4,9])
        bias = 0
        bias1 = 4
        bias2 = 18
        bias3 = 8
        bias4 = 17
        self.h1 = Neuron(weights, bias)
        self.h2 = Neuron(weights4, bias1)
        self.h3 = Neuron(weights5, bias2)
        self.h4 = Neuron(weights6,bias3)
        self.o1 = Neuron(weights3, bias4)
    def feedforward(self,x):
        out_h1 = self.h1.feedforward(x)
        out_h2 = self.h2.feedforward(x)
        out_h3 = self.h3.feedforward(x)
        out_h4 = self.h4.feedforward(x)
        out_o1 = self.o1.feedforward(np.array([out_h1,out_h2, out_h3, out_h4]))
        return out_o1

network1 = NeuralNet3()
x = np.array([2,3])
print(network1.feedforward(x)) 

```

```python
# Build a complete neural net including training
def deriv_sigmoid(x):
  # Derivative of sigmoid: f'(x) = f(x) * (1 - f(x))
  fx = sigmoid(x)
  return fx * (1 - fx)

def mse_loss(y_true, y_pred):
  # y_true and y_pred are numpy arrays of the same length.
  return ((y_true - y_pred) ** 2).mean()

class OurNeuralNetwork:
  
  def __init__(self):
    # Weights
    self.w1 = 1
    self.w2 = 3
    self.w3 = 12
    self.w4 = 0.5
    self.w5 = 2
    self.w6 = 9

    # Biases
    self.b1 = 0.7
    self.b2 = 17
    self.b3 = 0.2
    
  def feedforward(self, x):
    # x is a numpy array with 2 elements.
    h1 = sigmoid(self.w1 * x[0] + self.w2 * x[1] + self.b1)
    h2 = sigmoid(self.w3 * x[0] + self.w4 * x[1] + self.b2)
    o1 = sigmoid(self.w5 * h1 + self.w6 * h2 + self.b3)
    return o1

  def train(self, data, all_y_trues):
    
    learn_rate = 0.1
    epochs = 1000 # number of times to loop through the entire dataset

    for epoch in range(epochs):
      for x, y_true in zip(data, all_y_trues):
        # --- Do a feedforward (we'll need these values later)
        sum_h1 = self.w1 * x[0] + self.w2 * x[1] + self.b1
        h1 = sigmoid(sum_h1)

        sum_h2 = self.w3 * x[0] + self.w4 * x[1] + self.b2
        h2 = sigmoid(sum_h2)

        sum_o1 = self.w5 * h1 + self.w6 * h2 + self.b3
        o1 = sigmoid(sum_o1)
        y_pred = o1

        # --- Calculate partial derivatives.
        # --- Naming: d_L_d_w1 represents "partial L / partial w1"
        d_L_d_ypred = -2 * (y_true - y_pred)

        # Neuron o1
        d_ypred_d_w5 = h1 * deriv_sigmoid(sum_o1)
        d_ypred_d_w6 = h2 * deriv_sigmoid(sum_o1)
        d_ypred_d_b3 = deriv_sigmoid(sum_o1)

        d_ypred_d_h1 = self.w5 * deriv_sigmoid(sum_o1)
        d_ypred_d_h2 = self.w6 * deriv_sigmoid(sum_o1)

        # Neuron h1
        d_h1_d_w1 = x[0] * deriv_sigmoid(sum_h1)
        d_h1_d_w2 = x[1] * deriv_sigmoid(sum_h1)
        d_h1_d_b1 = deriv_sigmoid(sum_h1)

        # Neuron h2
        d_h2_d_w3 = x[0] * deriv_sigmoid(sum_h2)
        d_h2_d_w4 = x[1] * deriv_sigmoid(sum_h2)
        d_h2_d_b2 = deriv_sigmoid(sum_h2)

        # --- Update weights and biases
        # Neuron h1
        self.w1 -= learn_rate * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_w1
        self.w2 -= learn_rate * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_w2
        self.b1 -= learn_rate * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_b1

        # Neuron h2
        self.w3 -= learn_rate * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_w3
        self.w4 -= learn_rate * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_w4
        self.b2 -= learn_rate * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_b2

        # Neuron o1
        self.w5 -= learn_rate * d_L_d_ypred * d_ypred_d_w5
        self.w6 -= learn_rate * d_L_d_ypred * d_ypred_d_w6
        self.b3 -= learn_rate * d_L_d_ypred * d_ypred_d_b3

      # --- Calculate total loss at the end of each epoch
      if epoch % 10 == 0:
        y_preds = np.apply_along_axis(self.feedforward, 1, data)
        loss = mse_loss(all_y_trues, y_preds)
        print("Epoch %d loss: %.3f" % (epoch, loss))

# Define dataset
data = np.array([
  [-2, -1],  # Alice
  [25, 6],   # Bob
  [17, 4],   # Charlie
  [-15, -6], # Diana
])
all_y_trues = np.array([
  1, # Alice
  0, # Bob
  0, # Charlie
  1, # Diana
])

# Train our neural network!
network = OurNeuralNetwork()
network.train(data, all_y_trues)
```
